```python
# A line graph, years x-axis, gdp y-axis

plt.plot(years, gdp, color = 'green', marker = 'o')
plt.ylabel('Billions US $ ')
plt.xlabel('Years')
plt.show()
```

```python
# A bar chart 

plt.bar(years, gdp, 0.7,align='edge',color='green')
#plt.xticks(rotation=45)
plt.title('GDP of india from 1980 to 2017 in Billions US $')
plt.ticklabel_format(useOffset=False)
plt.show()
```

```python
# A Line chart

variance = [1,2,4,8,16,32,64]
bias_squared = [64,32,16,8,4,2,1]
total_error = [x+y for x,y in zip(variance, bias_squared)]
# X-axis
xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-', label='bias^2')
plt.plot(xs, total_error, 'b:', label='total error')

plt.legend(loc=9)
plt.xlabel('model complexity')
plt.title('The bias-variance tradeoff')
plt.show()
```
